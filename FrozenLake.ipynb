```python
def policy_evaluation(pi, P, gamma=1.0, theta=1e-10):

    """""
    Evaluates a given policy pi using a set of dynamics P.
    
    Params:
    - pi: A policy function.
    - P: A list of lists, where P[s][a] contains tuples of (probability, next_state, reward, done).
    - gamma: Discount factor (default is 1.0).
    - theta: A small threshold for determining convergence.
    
    Returns:
    - V: The state value function...
    """
    
    # Initialize value function for all states to 0. prev_V keeps track of the previous value estimates.
    prev_V = np.zeros(len(P))

    while True:  # Loop until convergence
        # Initialize a new array V
        V = np.zeros(len(P))  # Reset V to zero for each iteration
        for s in range(len(P)):  # Iterate over all states in the environment
            # Get the action prescribed by the current policy pi for state s
            for prob, next_state, reward, done in P[s][pi(s)]:
                # Update the value of state s based on the Bellman expectation equation
                # V[s] += prob * (reward + gamma * V(next_state))
                V[s] += prob * (reward + gamma * prev_V[next_state] * (not done))
                # - `prob` is the probability of transitioning to `next_state` from `s` by taking action `pi(s)`
                # - `reward` is the reward received when taking the action
                # - `prev_V[next_state]` is the current estimate of the value of the `next_state`
                # - `done` indicates if the next state is terminal (no future rewards if `done` is True)

        # Check if the maximum change between the old and new value function is below the threshold
        if np.max(np.abs(prev_V - V)) < theta:  # Convergence check based on the threshold `theta`
            break  # Exit the loop if the value function has converged
        
        # Update prev_V with the new estimates from V for the next iteration
        prev_V = V.copy()  # Copy the updated value function to continue the iterations

    return V  # Return the final converged value function
```

```python
def policy_improvement(V, P, gamma=1.0):
    Q = np.zeros((len(P), len(P[0])), dtype=np.float64)

    for s in range(len(P)):
        for a in range(len(P[s])):
            for prob, next_state, reward, done in P[s][a]:
                Q[s][a] += prob * (reward + gamma * V[next_state] * (not done)) # not greedifying but itterating instead 
    new_pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return new_pi
```

```python
def policy_iteration(P, gamma=1.0, theta = 1e-10):
    random_actions = np.random.choice(tuple(P[0].keys()), len(P))
    pi = lambda s: {s:a for s, a in enumerate(random_actions)}[s]

    while True:
        old_pi = {s:pi(s) for s in range(len(P))}
        V = policy_evaluation(pi, P, gamma, theta)
        pi = policy_improvement(V, P, gamma)

        if old_pi == {s:pi(s) for s in range(len(P))}: break
    return V, pi
```

```python
import numpy as np

def value_iteration(P, gamma=1.0, theta=1e-10):
    """
    Performs the value iteration algorithm for a given Markov Decision Process.
    
    :param P: The transition model of the MDP.
    :param gamma: Discount factor.
    :param theta: Threshold for determining convergence.
    :return: A tuple (V, pi) of the value function and policy.
    """
    V = np.zeros(len(P), dtype=np.float64)  # Initialize value function

    while True:
        Q = np.zeros((len(P), len(P[0])), dtype=np.float64)  # Initialize Q-function

        # Loop over all states
        for s in range(len(P)):
            # Loop over all actions
            for a in range(len(P[s])):
                # Loop over all possible next states
                for prob, next_state, reward, done in P[s][a]:
                    Q[s][a] += prob * (reward + gamma * V[next_state] * (not done))

        # Compute max Q for each state
        V_new = np.max(Q, axis=1)
        
        # Check for convergence
        if np.max(np.abs(V - V_new)) < theta:
            break

        V = V_new  # Update value function

    # Extract policy using the optimal value function
    pi = lambda s: {s: a for s, a in enumerate(np.argmax(Q, axis=1))}

    return V, pi
```

```python
import numpy as np
from tqdm import tqdm

def lin_dec_epsilon_greedy(env, init_epsilon=1.0, min_epsilon=0.01, decay_ratio=0.05, n_episodes=5000):
    # (a) Calculate the number of episodes over which to decay epsilon
    decay_episodes = n_episodes * decay_ratio

    # Name for the strategy
    name = 'Lin e-greedy {} {} {}'.format(init_epsilon, min_epsilon, decay_ratio)

    # Variables for calculating statistics (if needed)
    Q = np.zeros((env.action_space.n,))
    N = np.zeros((env.action_space.n,))
    Qe = np.empty((n_episodes, env.action_space.n))
    returns = np.empty(n_episodes)
    actions = np.empty(n_episodes, dtype=np.int)

    # Main loop for episodes
    for e in tqdm(range(n_episodes), desc='Episodes for ' + name, leave=False):
        # (3) Calculate the value of epsilon for the current episode
        epsilon = 1 - e / decay_episodes
        epsilon *= init_epsilon - min_epsilon
        epsilon += min_epsilon
        epsilon = np.clip(epsilon, min_epsilon, init_epsilon)

        # (4) Select action: epsilon-greedy strategy
        if np.random.random() > epsilon:
            action = np.argmax(Q)  # Exploitation
        else:
            action = np.random.randint(len(Q))  # Exploration

        # Execute the action in the environment
        _, reward, _, _ = env.step(action)

        # Update the counts and the Q-table
        N[action] += 1
        Q[action] = Q[action] + (reward - Q[action]) / N[action]

        # Store statistics
        Qe[e] = Q
        returns[e] = reward
        actions[e] = action

    # Return name and statistics
    return name, returns, Qe, actions
```

```python
import numpy as np
from tqdm import tqdm

def exp_dec_epsilon_greedy(env, init_epsilon=1.0, min_epsilon=0.01, decay_ratio=0.1, n_episodes=5000):
    # (a) Calculate the exponentially decaying epsilons
    decay_episodes = int(n_episodes * decay_ratio)
    rem_episodes = n_episodes - decay_episodes

    # Compute an array of epsilons, where epsilon decays exponentially
    epsilons = np.logspace(-2, 0, decay_episodes)
    epsilons *= init_epsilon - min_epsilon
    epsilons += min_epsilon
    epsilons = np.pad(epsilons, (0, rem_episodes), 'edge')

    # Name for the strategy
    name = 'Exp e-greedy {} {} {}'.format(init_epsilon, min_epsilon, decay_ratio)

    # Variables for calculating statistics (if needed)
    Q = np.zeros((env.action_space.n,))
    N = np.zeros((env.action_space.n,))
    Qe = np.empty((n_episodes, env.action_space.n))
    returns = np.empty(n_episodes)
    actions = np.empty(n_episodes, dtype=np.int)

    # Main loop for episodes
    for e in tqdm(range(n_episodes), desc='Episodes for ' + name, leave=False):
        # (3) Use the pre-computed epsilon for the current episode
        epsilon = epsilons[e]

        # (4) Select action: epsilon-greedy strategy
        if np.random.random() > epsilon:
            action = np.argmax(Q)  # Exploitation
        else:
            action = np.random.randint(len(Q))  # Exploration

        # Execute the action in the environment
        _, reward, _, _ = env.step(action)

        # Update the counts and the Q-table
        N[action] += 1
        Q[action] = Q[action] + (reward - Q[action]) / N[action]

        # Store statistics
        Qe[e] = Q
        returns[e] = reward
        actions[e] = action

    # Return name and statistics
    return name, returns, Qe, actions
```

```python
import numpy as np

class TenArmedBandit:
    def __init__(self):
        # 10 armed, initialize with random true action values
        self.q_true = np.random.normal(0, 1, 10)  #true action values
    
    def step(self, action): #pull function
        # Take an action (pull an arm) and return a reward from a normal distribution centered on the true value
        reward = np.random.normal(self.q_true[action], 1)  # reward based on arm's true value
        return reward
```

```python
import numpy as np
from tqdm import tqdm

def pure_exploitation(env, n_episodes=5000):
    Q = np.zeros(10)  # Q-values for each arm
    N = np.zeros(10)  # Count of times each arm was pulled
    rewards = np.zeros(n_episodes)

    for e in tqdm(range(n_episodes), desc='Pure Exploitation', leave=False):
        # Select action that maximizes Q-value (greedy)
        action = np.argmax(Q)
        
        # Take the action and get a reward
        reward = env.step(action)
        
        # Update Q-value and N
        N[action] += 1
        Q[action] += (reward - Q[action]) / N[action]
        
        # Store the reward
        rewards[e] = reward
    
    return rewards, Q

```

```python
def lin_dec_epsilon_greedy(env, init_epsilon=1.0, min_epsilon=0.01, decay_ratio=0.05, n_episodes=5000):
    decay_episodes = int(n_episodes * decay_ratio)
    epsilon = init_epsilon
    epsilon_decay = (init_epsilon - min_epsilon) / decay_episodes
    
    Q = np.zeros(10)
    N = np.zeros(10)
    rewards = np.zeros(n_episodes)
    
    for e in tqdm(range(n_episodes), desc='Linear Epsilon-Greedy', leave=False):
        # Decay epsilon
        epsilon = max(min_epsilon, epsilon - epsilon_decay)
        
        # Epsilon-greedy action selection
        if np.random.random() > epsilon:
            action = np.argmax(Q)
        else:
            action = np.random.randint(10)
        
        # Take action and get reward
        reward = env.step(action)
        
        # Update Q-value and N
        N[action] += 1
        Q[action] += (reward - Q[action]) / N[action]
        
        # Store the reward
        rewards[e] = reward
    
    return rewards, Q


```

```python
def exp_dec_epsilon_greedy(env, init_epsilon=1.0, min_epsilon=0.01, decay_ratio=0.1, n_episodes=5000):
    decay_episodes = int(n_episodes * decay_ratio)
    rem_episodes = n_episodes - decay_episodes
    
    epsilons = np.logspace(-2, 0, decay_episodes)
    epsilons *= init_epsilon - min_epsilon
    epsilons += min_epsilon
    epsilons = np.pad(epsilons, (0, rem_episodes), 'edge')
    
    Q = np.zeros(10)
    N = np.zeros(10)
    rewards = np.zeros(n_episodes)
    
    for e in tqdm(range(n_episodes), desc='Exponential Epsilon-Greedy', leave=False):
        # Get the pre-computed epsilon for this episode
        epsilon = epsilons[e]
        
        # Epsilon-greedy action selection
        if np.random.random() > epsilon:
            action = np.argmax(Q)
        else:
            action = np.random.randint(10)
        
        # Take action and get reward
        reward = env.step(action)
        
        # Update Q-value and N
        N[action] += 1
        Q[action] += (reward - Q[action]) / N[action]
        
        # Store the reward
        rewards[e] = reward
    
    return rewards, Q

```

```python
import numpy as np

def value_iteration(P, gamma=1.0, theta=1e-10):
    """
    Value Iteration algorithm to compute the optimal state-value function and policy.
    
    Parameters:
    P : list
        Transition probability matrix for the MDP.
        P[s][a] is a list of (probability, next_state, reward, done) tuples.
    gamma : float
        Discount factor.
    theta : float
        Convergence threshold.
    
    Returns:
    V : np.ndarray
        Optimal state-value function.
    pi : np.ndarray
        Optimal policy.
    """
    V = np.zeros(len(P), dtype=np.float64)  # (1) Initialize state-value function

    while True:  # (2) Loop until convergence
        Q = np.zeros((len(P), len(P[0])), dtype=np.float64)  # (3) Initialize action-value function

        for s in range(len(P)):  # (4) Iterate over all states
            for a in range(len(P[s])):  # Iterate over all actions
                for prob, next_state, reward, done in P[s][a]:  # (5) Iterate over transitions
                    # (6) Calculate the Q-value for this action
                    Q[s][a] += prob * (reward + gamma * V[next_state] * (not done))
        
        # (7) Check if the state-value function has converged
        if np.max(np.abs(V - np.max(Q, axis=1))) < theta:
            break
        
        # (8) Update the state-value function
        V = np.max(Q, axis=1)

    # (9) Extract the optimal policy
    pi = np.argmax(Q, axis=1)

    return V, pi

# Example usage:
# Define the MDP transition matrix P with format P[s][a] = [(prob, next_state, reward, done)]
# P = [[[...]], [[...]], ...]
# gamma = 0.99
# theta = 1e-6
# V, pi = value_iteration(P, gamma, theta)
```
