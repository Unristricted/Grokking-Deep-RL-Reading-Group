```python

def policy_evaluation(pi, P, gamma=1.0, theta=1e-10):
    """
    This is a full implementation of the policy-evaluation algorihtm. 
    All we need is the policy we're trying to evaluate and the MDP the policy runs on.
    The discount factor, gamma, defaults to 1 and theta is a small number that we use to check for convergence
    """
    
    # Initialize the first-iteration estimates of the state-value function to zero
    prev_V = np.zeros(len(P))

    while True:
        # Initialize the current-iteration elements to zero as well
        V = np.zeros(len(P))
        
        # Loop through all states to estimate the state-value funciton
        for s in range(len(P)):

            # We use the policy pi to get the possible transitions
            # And each transition tuple has a probablility, next state, reward, and a done flag
            for prob, next_state, reward, done in P[s][pi(s)]:

                # Calculate the value of that state by summing up the weighted value of that transition,
                # using the "done" flag to ensure the value of the next state is zero to avoid infinite sums
                V[s] += prob * (reward + gamma * prev_V[next_state] * (not done))

        # At the end of each iteration we make sure that the state-value functions are changing;
        # otherwise we call it converged
        if np.max(np.abs(prev_V - V)) < theta:
            break

        # Finally, "copy" to get ready for the next iteration
        # or return the latest state-value function
        prev_V = V.copy()
    return V
```

```python
def policy_improvement(V, P, gamma=1.0):
    """
    This algorithm takes the state-value function of the policy you want to improve (V), and the MDP (P). (and gamma too, optionally)
    """

    # Initialize the Q-function to zero
    Q = np.zeros((len(P), len(P[0])), dtype=np.float64)
    # Loop through the states, action, and transistions
    for s in range(len(P)):
        for a in range(len(P[s])):
            for prob, next_state, reward, done in P[s][a]:

                # We use these values to calculate the Q-function
                """Likely the most important line of code in the whole notebook"""
                Q[s][a] += prob * (reward + gamma * V[next_state] * (not done))
                
    # Finally, we obtain a new, greedy policy by taking the argmax of the Q-function of the original policy.
    new_pi = lambda s: {s:a for s, a in enumerate(np.argmax(Q, axis=1))}[s]
    return new_pi
```

```python
def policy_iteration(P, gamma=1.0, theta = 1e-10):
    """Policy iteration is pretty simple, and only needs the MDP"""
    random_actions = np.random.choice(tuple(P[0].keys()), len(P))

    # The first step is to create a randomly generated policy. Anything here should do.
    # Here we create a list of random actions and then map them to the states.
    pi = lambda s: {s:a for s, a in enumerate(random_actions)}[s]

    while True:
        # We keep a copy of the policy before modifying it
        old_pi = {s:pi(s) for s in range(len(P))}

        # Set the state-value function of the policy
        V = policy_evaluation(pi, P, gamma, theta)

        # Get an improved polocy
        pi = policy_improvement(V, P, gamma)
        
        # Check is the new policy is any different. If it's different, we do it all over again
        # If it's not, we break out and return an optimal policy and the optimal state-value function
        if old_pi == {s:pi(s) for s in range(len(P))}: break
            
    return V, pi
```
